# Extracting a pokemons learn set

This Jupyter notebook is for scaping data from Serebii.net. Specificly a pokemons pokemon yellow learn set

```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Important Note

Each pokemons page has a unique format, so you'll need to change the value inside of dfs in order to extract the correct data

```python
# Create an URL object
url = 'https://www.serebii.net/pokedex/074.shtml' # The pokemons page
# Create object page
page = requests.get(url)
```

```python
# reading all the tables as DF
dfs = pd.read_html(url)
```

```python
# Defining data set
lvl_up_set = dfs[12] # Change the value here if you're not getting the correct data, just jump around 
                     # till you find the correct table
lvl_up_set
```

```python
# Defining data set
tm_set = dfs[13]
tm_set
```

```python
# Cleaning data, starting with lvl up set
# This cell should have no errors if you're accessing the correct table
lvl_up_set = lvl_up_set.drop(0, axis=0)
header_row = 0
lvl_up_set.columns = lvl_up_set.iloc[header_row]
lvl_up_set = lvl_up_set.reset_index(drop=True)
lvl_up_set = lvl_up_set.drop(header_row)
lvl_up_set = lvl_up_set.loc[:, lvl_up_set.columns.notnull()]
lvl_up_set = lvl_up_set.rename(columns={'Att.': 'Power'})
lvl_up_set = lvl_up_set.iloc[::2]
lvl_up_set = lvl_up_set.reset_index(drop=True)
lvl_up_set = lvl_up_set.drop('Type',axis=1)
lvl_up_set['Power'] = lvl_up_set['Power'].str.replace('--','Status')
lvl_up_set['Level'] = lvl_up_set['Level'].replace('—',0)
lvl_up_set['Effect %'] = lvl_up_set['Effect %'].replace('--',0)
lvl_up_set = lvl_up_set.replace('--','Status', regex=True)
lvl_up_set = lvl_up_set.replace('_','0', regex=True)
lvl_up_set['Power'] = lvl_up_set['Power'].replace('??','Fixed')
lvl_up_set
```

```python
# attempitng to apply the same methods to TM moves
tm_set = tm_set.drop(0, axis=0)
header_row = 0
tm_set.columns = tm_set.iloc[header_row]
tm_set = tm_set.reset_index(drop=True)
tm_set = tm_set.drop(header_row)
tm_set = tm_set.loc[:, tm_set.columns.notnull()]
tm_set = tm_set.rename(columns={'Att.': 'Power'})
tm_set = tm_set.iloc[::2]
tm_set = tm_set.reset_index(drop=True)
tm_set = tm_set.drop('Type',axis=1)
tm_set['Power'] = tm_set['Power'].str.replace('--','Status')
tm_set['Power'] = tm_set['Power'].replace('??','Fixed')
tm_set['Effect %'] = tm_set['Effect %'].replace('--',0)
tm_set
```

```python
pokemons_name = dfs[1]
pokemons_name = pd.DataFrame(pokemons_name[0])
pokemons_name = pokemons_name[0][0].replace('#','')
pokemons_name
file_name_lvl_up = '{} lvl up set'.format(pokemons_name)
file_name_tm_set = '{} tm moves'.format(pokemons_name)
```

```python
lvl_up_set.to_csv(file_name_lvl_up)
```

```python
tm_set.to_csv(file_name_tm_set)
```
