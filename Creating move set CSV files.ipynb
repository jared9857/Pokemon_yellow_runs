# Extracting a pokemons learn set

This Jupyter notebook is for scaping data from Serebii.net. Specificly a pokemons pokemon yellow learn set

```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Important Note

Each pokemons page has a unique format, so you'll need to change the value inside of dfs in order to extract the correct data

```python
# Creating a list of numbers from 1-151

numbers = list(range(1,152))

# Turning the numbers to strings so they work with the URL

numbers_str = [str(int) for int in numbers]
print(numbers_str)
```

```python
# in order for the URL to work, the one and two digit numbers must lead with 0's so lets fix our list
temp_list = []
for number in numbers_str:
    temp_number = number.zfill(3)
    temp_list.append(temp_number)
print(temp_list)
```

```python
# Creating a list of all URL's

urls = []
for number in temp_list:
    url = 'https://www.serebii.net/pokedex/' + number +'.shtml'
    urls.append(url)
print(urls)
```

```python
# Due to the way my web scraper works and the way Serebii is structured I'll need to scrape differently depending on if the
# pokemon has a yellow specific move set, is onestage, or multi stage pokemon. 

# Starting with the Pokemon yellow specific movesets
yellow_lvl_up_pkm = []
for link in urls:
    url = link
    dfs = pd.read_html(url)
    try:
        if dfs[14].iloc[0][0] == 'Yellow Level Up':
            dataf = pd.DataFrame(dfs[14])
            dataf = dataf.drop(0, axis=0)
            dataf.columns = dataf.iloc[0]
            dataf = dataf.reset_index(drop=True)
            dataf = dataf.drop(0)
            dataf = dataf.loc[:, dataf.columns.notnull()]
            dataf = dataf.rename(columns={'Att.': 'Power'})
            dataf = dataf.iloc[::2]
            dataf = dataf.reset_index(drop=True)
            dataf = dataf.drop('Type',axis=1)
            dataf['Power'] = dataf['Power'].str.replace('--','Status')
            dataf['Level'] = dataf['Level'].replace('—',0)
            dataf['Effect %'] = dataf['Effect %'].replace('--',0)
            dataf = dataf.replace('--','Status', regex=True)
            dataf = dataf.replace('_','0', regex=True)
            dataf['Power'] = dataf['Power'].replace('??','Fixed')
            dataf['Name'] = dfs[5].iloc[1][0]
            print(pd.DataFrame(dataf))
            filename = dfs[5].iloc[1][0]
            yellow_lvl_up_pkm.append(filename)
            dataf.to_csv(f'C:\\Users\\jared\\data_science\\Pokemon_yellow_runs\\Yellow_movesets\\{filename}.csv')
    except:
        pass    
```

```python
# Creates the TM learnsets for the pokemon yellow learnset pokemon
for link in urls:
    url = link
    dfs = pd.read_html(url)
    for pkm in yellow_lvl_up_pkm:
        if pkm == dfs[5].iloc[1][0]:
            tm_set = pd.DataFrame(dfs[15])
            tm_set = tm_set.drop(0, axis=0)
            header_row = 0
            tm_set.columns = tm_set.iloc[header_row]
            tm_set = tm_set.reset_index(drop=True)
            tm_set = tm_set.drop(header_row)
            tm_set = tm_set.loc[:, tm_set.columns.notnull()]
            tm_set = tm_set.rename(columns={'Att.': 'Power'})
            tm_set = tm_set.iloc[::2]
            tm_set = tm_set.reset_index(drop=True)
            tm_set = tm_set.drop('Type',axis=1)
            tm_set['Power'] = tm_set['Power'].str.replace('--','Status')
            tm_set['Power'] = tm_set['Power'].replace('??','Fixed')
            tm_set['Effect %'] = tm_set['Effect %'].replace('--',0)
            tm_set['Name'] = dfs[5].iloc[1][0]
            filename = pkm
            print(tm_set.head())
            tm_set.to_csv(f'C:\\Users\\jared\\data_science\\Pokemon_yellow_runs\\Yellow_tms\\{filename}_tm.csv')
```
